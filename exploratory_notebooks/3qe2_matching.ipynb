```python
posA, posB = 0, 0
resA, resB = [], []
allresA = list(prot.get_residues())
allresB = list(rat_por.get_residues())
for a, b in zip(alignment.seqA, alignment.seqB):
    if a == b:
        resA.append(allresA[posA])
        resB.append(allresB[posB])
        posA += 1
        posB += 1
    elif a == "-":
        posB += 1
    elif b == "-":
        posA += 1
```
